```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
hd = pd.read_csv("F:\Data\Datasets\csv files\heart-disease.csv")
hd.head()
```

```python
over_50 = hd[hd["age"]>50]
len(over_50)
```

```python
#plotting subplots for 
# a> age VS cholesterol level
# b> age VS thalach
```

```python
fig , (ax0,ax1) = plt.subplots(nrows = 2,ncols = 1, figsize = (8,10))

#Plot1:  Age VS Cholesterol Level
scatter1 = ax0.scatter(x=over_50["age"],
                       y=over_50["chol"],
                       c=over_50["target"])
#adding labels and legends
ax0.set(title="Age VS Cholesterol Level",
        xlabel="Age",
       ylabel="Cholesterol")
ax0.legend(*scatter1.legend_elements(),title="target")

#adding mean-chol line

ax0.axhline(over_50["chol"].mean(),linestyle="--")



#Plot2:  Age VS thalach Level
scatter2 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"])
#adding labels and legends
ax1.set(title="Age VS Max Heart Rate",
        xlabel="Age",
       ylabel="Max Heart Rate")
ax1.legend(*scatter2.legend_elements(),title="target")

#adding mean-thalach line

ax1.axhline(over_50["thalach"].mean(),linestyle="--")

#adding title to the figure
fig.suptitle("Heart Disease Analysis", fontsize="15",fontweight="bold")
```

```python
fig , (ax0,ax1) = plt.subplots(nrows = 2,ncols = 1, figsize = (8,10))

#Plot1:  Age VS Cholesterol Level
scatter1 = ax0.scatter(x=over_50["age"],
                       y=over_50["chol"],
                       c=over_50["target"],
                        cmap="autumn")
#adding labels and legends
ax0.set(title="Age VS Cholesterol Level",
        xlabel="Age",
       ylabel="Cholesterol")
ax0.legend(*scatter1.legend_elements(),title="target")

#adding mean-chol line

ax0.axhline(over_50["chol"].mean(),linestyle="--")



#Plot2:  Age VS thalach Level
scatter2 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                       cmap="winter")
#adding labels and legends
ax1.set(title="Age VS Max Heart Rate",
        xlabel="Age",
       ylabel="Max Heart Rate")
ax1.legend(*scatter2.legend_elements(),title="target")

#adding mean-thalach line

ax1.axhline(over_50["thalach"].mean(),linestyle="--")

#adding title to the figure
fig.suptitle("Heart Disease Analysis", fontsize="15",fontweight="bold")


#setting x-limit and y-limit
ax0.set_xlim([50,80])
ax0.set_ylim([100,600])
```
