--- working over Heart Disease Dataset
- finding out the data of patients above 50 years of age.
+ here we compare 2 methods --
  + using a pyplot method
  + using matplotlib OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# calculating the length
len(over_50)
```

1. Pyplot method 

```python
over_50.plot(kind = "scatter",
             x = "age",
             y="chol",
             c="target",
            figsize=(10,5))
```

2. Matplotlib OO Method

```python
fig , ax = plt.subplots(figsize=(10,5))
over_50.plot(kind="scatter",
                x = "age",
                 y="chol",
                 c="target",
               ax=ax)
ax.set_xlim([45,100])
```

```python
#using Matplotlib OO Method using Object-oriented technique

fig, ax = plt.subplots(figsize = (10,5))

scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

#setting labels
ax.set(title = "Age VS Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol Level")

#adding legend
ax.legend(*scatter.legend_elements(),title="Target")

#adding hz line
ax.axhline(over_50["chol"].mean(),linestyle='--',dash_joinstyle="bevel")
```
